```python
from twitterscraper import query_tweets
from os import path
import datetime as dt
import pandas as pd
```

```python
# Twitterscraper code: provided an incomplete output of tweets
all_tweets = query_tweets("from:realDonaldTrump ", begindate=dt.date(2012,1,1), enddate=dt.date.today(), limit = None)
tweets_list = []
l = []
for i in all_tweets:
    l += [{"created_at" : i.timestamp, "text" : i.text}]
scraper = pd.DataFrame(l)
print("Done")
    
if path.exists("trump_tweet_scraper.csv"):
    scraper = pd.read_csv("trump_tweet_scraper.csv", engine='python', index_col = 0)
else:
    scraper.to_csv("trump_tweet_scraper.csv")
scraper
```
